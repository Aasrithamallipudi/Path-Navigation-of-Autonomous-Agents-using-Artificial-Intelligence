```python
import heapq
import numpy as np
import matplotlib.pyplot as plt

# Define the full graph with step costs
graph = {
    "Arad": {"Sibiu": 140, "Timisoara": 118, "Zerind": 75},
    "Sibiu": {"Arad": 140, "Fagaras": 99, "Oradea": 151, "Rimnicu Vilcea": 80},
    "Fagaras": {"Sibiu": 99, "Bucharest": 211},
    "Rimnicu Vilcea": {"Sibiu": 80, "Pitesti": 97, "Craiova": 146},
    "Pitesti": {"Rimnicu Vilcea": 97, "Bucharest": 101, "Craiova": 138},
    "Craiova": {"Rimnicu Vilcea": 146, "Pitesti": 138, "Dobreta": 120},
    "Timisoara": {"Arad": 118, "Lugoj": 111},
    "Lugoj": {"Timisoara": 111, "Mehadia": 70},
    "Mehadia": {"Lugoj": 70, "Dobreta": 75},
    "Dobreta": {"Mehadia": 75, "Craiova": 120},
    "Oradea": {"Sibiu": 151, "Zerind": 71},
    "Zerind": {"Arad": 75, "Oradea": 71},
    "Bucharest": {"Fagaras": 211, "Pitesti": 101, "Giurgiu": 90, "Urziceni": 85},
    "Giurgiu": {"Bucharest": 90},
    "Urziceni": {"Bucharest": 85, "Hirsova": 98, "Vaslui": 142},
    "Hirsova": {"Urziceni": 98, "Eforie": 86},
    "Eforie": {"Hirsova": 86},
    "Vaslui": {"Urziceni": 142, "Iasi": 92},
    "Iasi": {"Vaslui": 92, "Neamt": 87},
    "Neamt": {"Iasi": 87}
}

# Define heuristic values (straight-line distances to Bucharest)
heuristic = {
    "Arad": 366, "Bucharest": 0, "Craiova": 160, "Dobreta": 242, "Eforie": 161,
    "Fagaras": 178, "Giurgiu": 77, "Hirsova": 151, "Iasi": 226, "Lugoj": 244,
    "Mehadia": 241, "Neamt": 234, "Oradea": 380, "Pitesti": 98, "Rimnicu Vilcea": 193,
    "Sibiu": 253, "Timisoara": 329, "Urziceni": 80, "Vaslui": 199, "Zerind": 374
}

# A* algorithm for shortest path search
def astar_search(graph, heuristic, start, goal):
    queue = [(heuristic[start], 0, start, [start])]
    visited = set()

    while queue:
        f, g, node, path = heapq.heappop(queue)

        if node in visited:
            continue
        visited.add(node)

        if node == goal:
            return path, g

        for neighbor, step_cost in graph[node].items():
            if neighbor not in visited:
                g_new = g + step_cost
                h_new = heuristic[neighbor]
                f_new = g_new + h_new
                heapq.heappush(queue, (f_new, g_new, neighbor, path + [neighbor]))

    return None, float("inf")

# Simulate motion parameters (velocity, acceleration, rotation rate)
def simulate_motion(path):
    velocities, accelerations, rotations = [], [], []

    for i in range(len(path) - 1):
        distance = graph[path[i]][path[i + 1]]
        velocity = np.random.randint(5, 15)  # Random velocity between 5-15 m/s
        acceleration = np.random.uniform(0.1, 2.0)  # Random acceleration
        rotation_rate = np.random.uniform(-30, 30)  # Random rotation angle in degrees

        velocities.append(velocity)
        accelerations.append(acceleration)
        rotations.append(rotation_rate)

    return velocities, accelerations, rotations

# Visualize path with velocity, acceleration, rotation
def visualize_motion(path, velocities, accelerations, rotations):
    plt.figure(figsize=(12, 6))

    plt.plot(range(len(path) - 1), velocities, label="Velocity (m/s)", marker='o')
    plt.plot(range(len(path) - 1), accelerations, label="Acceleration (m/s)", marker='s')
    plt.plot(range(len(path) - 1), rotations, label="Rotation Rate ()", marker='x')

    plt.xlabel("Path Transitions")
    plt.ylabel("Values")
    plt.title("Robot Motion Planning Parameters Along the Path")
    plt.legend()
    plt.grid()
    plt.show()

# Execution
start, goal = "Arad", "Bucharest"
path, cost = astar_search(graph, heuristic, start, goal)

if path:
    print(f"Path: {' -> '.join(path)}, Total Cost: {cost}")
    velocities, accelerations, rotations = simulate_motion(path)
    visualize_motion(path, velocities, accelerations, rotations)
else:
    print("No path found!")
```
